# 열의 데이터를 분리(str 형태)

```python
item_to_choice = data[['item_name','choice_description','quantity']]
item_to_choice['choice_description'] = item_to_choice['choice_description'].apply(lambda x : ','.join(x.replace('[','').replace(']','').split(',')) if type(x) is str else 'None')
```

```python
item_to_choice
```

```python
item_to_choice = item_to_choice.reset_index()
```

```python
item_to_choice.columns = ['id','item_name','choice_description','quantity']
```

```python
choice_split = pd.DataFrame(item_to_choice.choice_description.str.split(',').tolist(),index=item_to_choice.id).stack()
```

```python
choice_split
```

```python
choice_split = choice_split.reset_index([0,'id'])
```

```python
choice_split
```

```python
result = pd.merge(item_to_choice[['id','item_name','quantity']],choice_split,how='left',on='id')
```

```python
result.columns = ['id','item_name','quantity','choice']
```

```python
del result['id']
result
```

# 각 그룹별 최댓값 찾기(multi index)

```python
get_top_by_item_name = result.groupby(['item_name','choice']).sum().sort_values('quantity').groupby(level=0).tail(1).sort_index()
```

```python
get_top_by_item_name.sort_values("quantity",ascending=False)[:10]
```
